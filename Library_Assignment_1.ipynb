Question 1: Rice Market Analysis

```python
import numpy as np



prices =np.array([1200, 1250, 1180, 1300, 1220, 1270, 1240])
print("Array:" , prices)

#b) Mean Price
mean_prices=np.mean(prices)
print(f"Mean_Prices: {mean_prices: .2f} LRD")

#c) Highest and lowest
print(f"Highest: {np.max(prices)} LRD | Lowest: {np.min(prices)} LRD")

# Difference
print(f"Difference: {np.max(prices) - np.min(prices)} LRD")

#e) The mean (1237.14 LRD) represents the typical daily price
#It gives the seller a central reference point to judge weaather
# any given day's price is above or below normal, and to estimate
# expected weekly revenue (1237.14 x 7 = ~8,660 LRD)


# Question 2

prices_shock =np.array([1200, 1250, 1180, 1300, 1220, 1270, 2000])
print(f"Q2a). Array_New:" , prices_shock)

new_mean =np.mean(prices_shock)
print(f"Q2b). New_means: {new_mean: .2f} LRD")

#b) Compare
print(f"Old Mean: {mean_prices:.2f} LRD | New_mean: {new_mean: .2f} LRD")
print(f"Difference: +{new_mean - mean_prices: 2f} LRD")

#c) why outliers affect analysis
# One usually high value (2000 LRD) raised the mean by ~108 LRD, give
# a false picture that prices are generally higher.
# The mean is sensitive to outliers | a single extreme value
# can distort it significantly, misleading buyers and sellers.
```
