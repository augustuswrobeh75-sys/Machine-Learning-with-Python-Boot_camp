```python
import numpy as np
 # Daily gari prices at Gobachop Market (LRD per bag)
prices = np.array([850, 900, 875, 920, 860, 910, 895])

print('prices:' ,        prices)
print('Mean price:' ,    np.mean(prices))
print('Median price' ,   np.median(prices))
print('Highest:'  ,      np.max(prices))
print('Lowest:' ,        np.min(prices))
print('Std Dev'  ,      np.std(prices).round(2))
```

```python
import pandas as pd

# Open the ledger book
url="https://raw.githubusercontent.com/augustuswrobeh75-sys/Machine-Learning-with-Python-Boot_camp/refs/heads/main/gari_price_data.csv"
df = pd.read_csv(url)


print(df.head())              # first 5 rows
print( '\nShape:' , df .shape)      # rows x columns
print('\nColumns:' , df.columns.tolist())
print('\nDescribe:\n' , df.describe())
print(df.tail())
print(df.iloc[11:25 ,:])

#select one column
prices = df['price_next_day_lrd']
print(prices.head())

# filter rows where rainfall is heavy
rainy_days =df[df['rainfall_mm'] > 10]
print(f'Rainy days: {len(rainy_days)} rows')
print(rainy_days [['rainfall_mm', 'price_next_day_lrd']].head())


# filter rows where rainfall is heavy
rainy_days =df[df['rainfall_mm'] > 10 ]
print(f'Rainy days: {len(rainy_days)} rows')
print(rainy_days [['rainfall_mm', 'price_next_day_lrd']].head())
```

```python
import matplotlib.pyplot as plt
days  = list(range(1, 8))
prices = [850, 900, 875, 920, 860, 910, 895]

plt.figure(figsize=(8,4))
plt.plot(days, prices, marker='o' , color='darkorange', linewidth=2)
plt.title('Gari Price Trend - Gobacho Market (LRD)')
plt.xlabel('Day') ; plt.ylabel('Price per Bag (LRD)')
plt.grid(True)  ; plt.tight_layout()    ; plt.show()
```
